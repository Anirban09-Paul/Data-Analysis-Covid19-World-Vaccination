# check how many null value has in each coulmns

```python
df.isnull().sum()
```

# convert to percent 

```python
df_per=df.isnull().sum()/df.shape[0]*100
df_per
```

now again check if a coulmn has less than 50 percent null value then we will take it all either through...

```python
missing_value=df_per[df_per>50].keys()
missing_value
```

drop the values greater than 50 percent

```python
df2=df.drop(columns=missing_value)
df2.shape
```

```python
df2.select_dtypes(include=['float64'])
```

# To check how many null values in visualizing mode...

```python
plt.figure(figsize=(8,5))
sns.heatmap(df2.isnull())
```

# now i am taking mean, median, mode to fill the missing values

```python
df3=df2.fillna(df2.mean())
```
