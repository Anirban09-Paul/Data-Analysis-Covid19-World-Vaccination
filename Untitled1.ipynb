```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('country_vaccinations.csv')
df
```

```python
df.isnull().sum().sum()
```

```python
df.info()
```

```python
missing_value=df.isnull().sum()/df.shape[0]*100
missing_value.head()
```

```python
find_missing_value=missing_value[missing_value>50].keys()
find_missing_value
```

```python
drop_clms=df.drop(columns=find_missing_value)
drop_clms.shape
```

```python
df2=drop_clms.select_dtypes(include=['float64'])
df2
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(df2.isnull())
```

```python
df3=df2.fillna(df2.mean())
# df4=df2.fillna(df2.median())
df3.head()
# df4.head()
```
